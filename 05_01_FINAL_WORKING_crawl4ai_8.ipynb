```python
# Read the file
with open("05_01_FINAL_WORKING_crawl4ai_JUPYTER.py", "r") as f:
    lines = f.readlines()

# Find where "if __name__" starts and exclude everything after it
script_lines = []
for line in lines:
    if "if __name__" in line:
        break
    script_lines.append(line)

# Execute only the function definitions
exec("".join(script_lines))

# Now call the function with await (Jupyter-friendly)
data_res = await crawl_website()
print(f"Crawled {len(data_res['data'])} pages")
```

```python
import subprocess
import sys
import json

# Run your script
result = subprocess.run(
    [sys.executable, "path/to/your_script.py"],
    capture_output=True,
    text=True,
    timeout=120,
)

if result.returncode == 0:
    # Parse the output
    lines = result.stdout.strip().split("\n")
    json_line = lines[-1]  # Get the last line

    data_res = json.loads(json_line)
    print("Successfully crawled!")
    print(f"Number of results: {len(data_res['data'])}")
else:
    print("Error:", result.stderr)
```
